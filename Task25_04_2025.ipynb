**IMPORTING REQURING LIBRARY**



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
sns.set(style='whitegrid')
```

**IMPORTING** **DATASET**

```python
data = pd.read_csv("/content/Titanic-Dataset.csv")
```

**STATISTICS SUMMARY**

```python
data.describe()
```

**HISTOGRAM AND BOXPLOT FOR NUMERICAL FEATURE**

```python
#histogram for numeric feature
data.hist(bins=20, figsize=(10,8))
plt.tight_layout()
plt.show()

#boxplot for numeric feature
data.plot(kind='box', subplots=True, layout=(1,len(data.select_dtypes(include='number').columns)),figsize=(12,6), sharex=False)
plt.tight_layout()
plt.show()
```

**CORRELATION MATRIX**

```python
# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**PATTERNS AND TRENDS**

```python
df = data[['Survived', 'Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch', 'Embarked']]


# Boxplot: Age vs Survived (Outliers + distribution)
plt.figure(figsize=(8, 5))

sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()

# Boxplot: Fare vs Survived (check fare outliers)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.show()

# Histograms: Age and Fare

df[['Age', 'Fare']].hist(bins=30, figsize=(10, 4), layout=(1, 2))
plt.suptitle('Distribution of Age and Fare')
plt.show()
```

**FEATURE LEVEL INFERENCE**

```python
# Countplot: Survival by Sex

sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender')
plt.show()

# Countplot: Survival by Class

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()

# Pairplot: Visualizing multiple relationships

sns.pairplot(df.dropna(), hue='Survived', vars=['Age', 'Fare', 'SibSp', 'Parch'])
plt.suptitle('Pairplot of Numeric Features by Survival', y=1.02)
plt.show()

# Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```
